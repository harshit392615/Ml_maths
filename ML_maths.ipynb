Basic Statistics

Mean

```python
def mean(v):
  sum = 0
  for i in v:
    sum += i
  return sum/len(v)
x_mean = mean(x)
y_mean = mean(y)

print(x_mean)
print(y_mean)
```

Mean Centering

We often subtract the mean from each value:

Shifts data so its mean becomes 0

Does not change relationships between variables

Makes slope, covariance, and gradient calculations cleaner

```python
def mean_centering(v , v_mean):
  return v-v_mean

x_mc = mean_centering(x , x_mean)
y_mc = mean_centering(y , y_mean)

print(x_mc , y_mc)
```

Variance

```python
def variance(v_mc):
  sum = 0
  for v in v_mc:
    sum += v**2
  return sum/len(v_mc)

var_x = variance(x_mc)
var_y = variance(y_mc)

print(var_x)
print(var_y)
```

Covariance

```python
def covariance(v1_mc , v2_mc):
  sum = 0
  i = 0
  while i<len(v1_mc):
    sum += v1_mc[i]*v2_mc[i]
    i+=1
  return sum/len(v1_mc)

covar = covariance(x_mc , y_mc)

print(covar)
```
