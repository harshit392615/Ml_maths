Feature scaling

```python
import numpy as np
x = np.array([1.2, 2.0, 2.8, 3.6, 4.5, 5.1, 6.0, 6.7])
y = np.array([8.5, 11.1, 13.9, 16.2, 19.8, 21.4, 23.9, 26.1])
```

Standardization (Z-score scaling)

```python
mean = x.mean()
std = x.std()
if std == 0 :
  std = 1

x_scaled = (x-mean)/std
print(x_scaled)
```

Normalization (Min-Max scaling)

```python
max = x.max()
min = x.min()

x_scaled = (x-min)/(max-min)
print(x_scaled)
```
